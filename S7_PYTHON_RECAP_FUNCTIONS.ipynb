*En esta sección vamos a ver como tratar los valores faltantes. Existen 3 tipos:*

* __MCAR (MISSING COMPLETLY AT RANDOM):__ Cada registro que es nulo lo hace sin ningún patron específico. Es decir, cualquiera de los registros tiene la misma chance de estar vacío. RECOMENDACIÓN: ELIMINAR O IMPUTAR (RELLENAR VALORES CON MEDIA O LA MEDIANA)

* __MAR (MISSING AT RANDOM):__ Hay valores nulos, pero estos se relacionan con otra columna específica. Es decir, hay valores nulos, pero estos se asocian a un unico o unicos. Por ejemplo: Quantity esta vacio pero esta relacionado con el mismo country.

* __MNAR (MISSING AT NO RANDOM):__ El valor no está o aparece nulo, pero no es al azar. Por ejemplo: Persona de áreas ricas de la ciudad, no dan su nombre por privacidad, por lo que la coluna user_name, suele tener nulos.

```python
def text_column_check(df, column_name):
  df[column_name] = df[column_name].astype('str')       #SUB-PROCESO 1: CONVERTIR AL TIPO DE DATOS ADECUADO.
  df[column_name] = df[column_name].str.strip()         #SUB-PROCESO 2: ELIMINAR ESPACIOS EN BLANCO.
  df[column_name] = df[column_name].str.lower()         #SUB-PROCESO 3: DEJAR TODO TEXTO EN MAYÚSCULAS.
  df[column_name].fillna('DESCONOCIDO', inplace = True) #SUB-PROCESO 4: IMPUTAR COLUMNA CON "DESCONOCIDO"
  df[column_name] = df[column_name].replace("nan","DESCONOCIDO") #SUB-PROCESO 5: QUITAR PALABRA nan Y SUSTUIRLA.
  return df
```

```python
def numeric_column_check(df,column_name):
  df[column_name] = pd.to_numeric(df[column_name], errors='coerce') #SUB-PROCESO 1: DEFINIR COMO NUMÉRICO.
  median_column   = df[column_name].median()
  df[column_name].fillna(median_column, inplace = True)
  return df
```

```python
df_column_imputation_1 = numeric_column_check(df = df_work,column_name = 'price_per_unit')
```

```python
df_column_imputation_1.isna().sum()
```

```python
lista_numeric = ["total_weight","shipping_cost","tax_amount"]
```
